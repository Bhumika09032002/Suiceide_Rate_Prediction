```python
dummies1=pd.get_dummies(data.CAUSE)
dummies1
```

```python
dummies2=pd.get_dummies(data['STATE/UT'])
dummies2
```

```python
data_new=pd.concat([data,dummies1,dummies2],axis='columns')
data_new
```

```python
data_new=data_new.drop('STATE/UT',axis='columns')
data_new.head()
```

```python
data_new=data_new.drop('CAUSE',axis='columns')
data_new.head()
```

```python
data_new=data_new.drop('Total Male',axis='columns')
data_new.head()
```

```python
data_new=data_new.drop('Total Female',axis='columns')
data_new.head()
```

```python
x=data_new.drop('Total',axis='columns')
x.head()
```

```python
y=data_new.Total
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
print(x_test)
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
import warnings
warnings.simplefilter('ignore')

poly=PolynomialFeatures(degree=2)
X=poly.fit_transform(x)
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
regressor=DecisionTreeRegressor(random_state=0)
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
knn_model = KNeighborsRegressor(n_neighbors=3)
knn_model.fit(x_train,y_train)
```

```python
knn_model.score(x_test,y_test)
```

```python
from sklearn.svm import SVR
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
sv=SVR(kernel='rbf')
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)
```

```python
from sklearn.linear_model import Ridge
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
r = Ridge(alpha=1.0)
r.fit(x_train, y_train)
```

```python
r.score(x_test,y_test)*100
```

```python
from sklearn import linear_model
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
logr = linear_model.LogisticRegression()
logr.fit(x_train,y_train)
```

```python
logr.score(x_test,y_test)*100
```

```python
from sklearn.linear_model import BayesianRidge
model = BayesianRidge()
model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import Lasso
lassomodel = Lasso(alpha=1.0)
lassomodel.fit(x_train, y_train)
```

```python
lassomodel.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
regressor.fit(x_train, y_train)  
```

```python
regressor.score(x_test,y_test)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
```

```python
kernel = 1 * RBF(length_scale=1.0, length_scale_bounds=(1e-2, 1e2))
gaussian_process = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=9)
```

```python
gaussian_process.fit(x_train, y_train)
```

```python
gaussian_process.score(x_test,y_test)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC(kernel='rbf') # Linear Kernel
clf.fit(x_train, y_train)
```

```python
clf.score(x_test,y_test)
```
